```python
from selenium import webdriver
import pandas as pd
import time
browser = webdriver.Chrome()

# Opening the url using selenium
# Reading csv with pandas
df = pd.read_csv('C:\\Users\\DELL\\Downloads\\Untitled spreadsheet - hotels_data greaternoida (1).csv')  # Replace 'file.csv' with the actual filename and path of your csv
names = df['title']  # Assuming you want to extract a column called 'title'

# List to store current URLs
current_urls = []

for i in names:
    url = f'https://www.google.com/maps/search/{i}'
    browser.get(url)
    time.sleep(5)
    current_url = browser.current_url
    current_urls.append(current_url)

# Add the current URLs to the original DataFrame
df['current_url'] = current_urls

# Save the DataFrame back to the original CSV file
df.to_csv('C:\\Users\\DELL\\Downloads\\Untitled spreadsheet - hotels_data greaternoida (1).csv', index=False)

# Close the browser when done
browser.quit()
```

```python
import csv
import re
import pandas as pd

# Function to extract latitude and longitude from a Google Maps link
def extract_lat_long(url):
    pattern = r'@(-?\d+\.\d+),(-?\d+\.\d+)'
    match = re.search(pattern, url)
    if match:
        latitude = match.group(1)
        longitude = match.group(2)
        return latitude, longitude
    else:
        return None, None

# Read the CSV file into a DataFrame
csv_file_path = 'C:\\Users\\DELL\\Downloads\\Untitled spreadsheet - hotels_data greaternoida (1).csv'  # Replace 'your_file.csv' with the actual file path
df = pd.read_csv(csv_file_path)

# Apply the extract_lat_long function to each URL and create new columns for latitude and longitude
df[['Latitude', 'Longitude']] = df['current_url'].apply(lambda x: pd.Series(extract_lat_long(x)))

# Save the modified DataFrame back to the CSV file
df.to_csv(csv_file_path, index=False)

print("Latitude and longitude added to the CSV file.")
```

```python
import folium
import pandas as pd

# Read the CSV file into a DataFrame
csv_file_path = 'C:\\Users\\DELL\\Downloads\\HR botics\\imp\\New folder\\rest_GN.csv'
df = pd.read_csv(csv_file_path)

# Create a map centered around the first location in the DataFrame
map_center = [df['Latitude'].iloc[0], df['Longitude'].iloc[0]]
my_map = folium.Map(location=map_center, zoom_start=12)

# Add markers for each location in the DataFrame with red marker style
for index, row in df.iterrows():
    hotel_name = f"Hotel Name: <span style='font-size: 14px;'>{row['title']}</span>"  # Include "Hotel Name:" in front of the title with adjusted font size
    location = [row['Latitude'], row['Longitude']]
    
    # Create HTML for the thumbnails with max-height property
    thumbnails_html = ''.join([f'<img src="{thumbnail}" style="max-height:100px; max-width:200px; margin-right:5px;">' for thumbnail in eval(row['thumbnails'])])
    
    # Use HTML formatting for the popup text with thumbnails, reviews, and set marker to red style
    popup_text = f"{hotel_name}<br>Rating: <span style='font-size: 14px; color: yellow;'>&#9733; {row['rating']}</span><br>Reviews: <span style='font-size: 14px;'>{row['reviews']}</span><br>Price: <span style='font-size: 14px;'>{row['Price']}</span><br>{thumbnails_html}"
    
    # Set the marker to red style and add the price to the marker itself
    marker = folium.Marker(location=location, popup=popup_text, icon=folium.Icon(color='red'), tooltip=f"Price: {row['Price']}", parse_html=True)
    marker.add_to(my_map)

# Save the map to an HTML file
my_map.save("Greater_noida_restaurants.html")
```
