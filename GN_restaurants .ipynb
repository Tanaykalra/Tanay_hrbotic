```python
import folium
import pandas as pd

# Load the data from the CSV file
data_path = r'C:\Users\DELL\Downloads\HR botics\imp\cleaned_data.csv'
df = pd.read_csv(data_path)

# Create a Folium map centered around the average coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
map_osm = folium.Map(location=map_center, zoom_start=14)

# Add markers for each data point
for _, row in df.iterrows():
    popup_text = f"{row['Name']} (Rating: {row['Rating']})"
    folium.Marker(location=[row['Latitude'], row['Longitude']],
                  popup=popup_text).add_to(map_osm)

# Save the map to an HTML file
map_osm.save('restaurant_map.html')
```

```python
import folium
import pandas as pd

# Load the data from the CSV file
data_path = r'C:\Users\DELL\Downloads\GNrestaurants_cleaned .csv'
df = pd.read_csv(data_path)

# Create a Folium map centered around the average coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
map_osm = folium.Map(location=map_center, zoom_start=14, tiles='CartoDB positron')

# Add markers for each data point
for _, row in df.iterrows():
    # Add a star symbol in front of the ratings
    rating_with_star = f"â˜… {row['Rating']}"
    
    # HTML for the popup, including the image
    popup_html = f"""
    <div>
        <h4>Name: {row['Name']}</h4>
        <p style="font-size:16px;">Rating: {rating_with_star}</p>
        <img src="{row['attributes']}" alt="Restaurant Image" style="max-width:200px;">
    </div>
    """

    # Create a Popup with the HTML content
    popup = folium.Popup(html=popup_html, max_width=300)

    # Create the Marker with the Popup
    folium.Marker(location=[row['Latitude'], row['Longitude']],
                  popup=popup).add_to(map_osm)

# Save the map to an HTML file
map_osm.save('restaurant_map_with_images_and_stars.html')
```
