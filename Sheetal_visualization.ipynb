```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a DataFrame with the provided data
data = {
    'Brands': ['SHEETAL', 'GANGA', 'SINTEX', 'SUPREME', 'ASHIRWAD', 'APPOLO', 'ASTRAL'],
    'BRAND RATINGS': [3.6, 3.8, 4.5, 4.43, 4.20, 3.83, 4.43],
    'Google Ratings': [3, 3.5, 4.7, 4.4, 4.4, 3.8, 4.7],
    'Just Dial Ratings': [4.2, 4.3, 4.4, 4.5, 4.5, 4.1, 4.5],
    'Indiamart Ratings': [4.2, 4, 4.2, 4.4, 3.7, 3.6, 4.1],
    'Amazon Ratings': [2.8, 3.4, 4.5, 0, 0, 0, 0],
    'Average Rating': [3.55, 3.80, 4.45, 3.55, 3.36, 3.07, 3.55]
}

df = pd.DataFrame(data)

# Setting up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
fig.suptitle('Ratings Comparison Across Brands and Platforms', fontsize=16)

# Define vibrant colors for each brand
brand_colors = {'SHEETAL': '#FF6F61', 'GANGA': '#5E60CE', 'SINTEX': '#FFB74D', 
                'SUPREME': '#4CAF50', 'ASHIRWAD': '#FFD700', 'APPOLO': '#8A2BE2', 'ASTRAL': '#FF69B4'}

# Plotting each metric on subplots with vibrant colors and annotations
sns.barplot(x='BRAND RATINGS', y='Brands', data=df, ax=axes[0, 0], palette=brand_colors)
axes[0, 0].set_title('BRAND RATINGS')
for i, v in enumerate(df['BRAND RATINGS']):
    axes[0, 0].text(v + 0.05, i, f'{v}', color='black', va='center', fontweight='bold')

sns.barplot(x='Google Ratings', y='Brands', data=df, ax=axes[0, 1], palette=brand_colors)
axes[0, 1].set_title('Google Ratings')
for i, v in enumerate(df['Google Ratings']):
    axes[0, 1].text(v + 0.05, i, f'{v}', color='black', va='center', fontweight='bold')

sns.barplot(x='Just Dial Ratings', y='Brands', data=df, ax=axes[1, 0], palette=brand_colors)
axes[1, 0].set_title('Just Dial Ratings')
for i, v in enumerate(df['Just Dial Ratings']):
    axes[1, 0].text(v + 0.05, i, f'{v}', color='black', va='center', fontweight='bold')

sns.barplot(x='Indiamart Ratings', y='Brands', data=df, ax=axes[1, 1], palette=brand_colors)
axes[1, 1].set_title('Indiamart Ratings')
for i, v in enumerate(df['Indiamart Ratings']):
    axes[1, 1].text(v + 0.05, i, f'{v}', color='black', va='center', fontweight='bold')

# Adjusting layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Display the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
brands = ["Sintex Water Tanks", "Supreme Water Tanks", "Vectus Water Tanks", 
          "Ocean Water Tanks", "Plasto Water Tanks", "Sheetal Water Tanks", 
          "National Plastic Water Tanks"]
revenue = [3203.86, 1800, 816, 600, 520, 270, 112]

# Create a DataFrame
data = {"Brands": brands, "Revenue (Cr)": revenue}
df = pd.DataFrame(data)

# Plotting with Seaborn color palette
sns.set_palette("Accent")
plt.figure(figsize=(8, 8))

# Pie chart
plt.pie(df["Revenue (Cr)"], labels=df["Brands"], autopct='%1.1f%%', startangle=180)
plt.title('Water Tank Brands Revenue Distribution')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    'SUPPLY CHAIN (Number)': ['Ghaziabad', 'Gr. Noida', 'Noida', 'Faridabad', 'Total of all 4 cities'],
    'SHEETAL': [17, 8, 11, 8, 44],
    'GANGA': [33, 27, 29, 10, 99],
    'SINTEX': [11, 6, 25, 9, 51],
    'SUPREME': [5, 11, 24, 5, 45],
    'ASHIRWAD': [6, 2, 0, 2, 10],
    'APPOLO': [4, 0, 0, 0, 4],
    'ASTRAL': [9, 0, 8, 1, 18]
}

df = pd.DataFrame(data)

# Set a dark color palette for better visibility
sns.set_palette("summer")

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
fig.suptitle('Supply Chain Distribution Across cities ', fontsize=16)

# Plotting each brand in subplots
sns.barplot(x='SUPPLY CHAIN (Number)', y='SHEETAL', data=df, ax=axes[0, 0])
axes[0, 0].set_title('SHEETAL')

sns.barplot(x='SUPPLY CHAIN (Number)', y='GANGA', data=df, ax=axes[0, 1])
axes[0, 1].set_title('GANGA')

sns.barplot(x='SUPPLY CHAIN (Number)', y='SINTEX', data=df, ax=axes[1, 0])
axes[1, 0].set_title('SINTEX')

sns.barplot(x='SUPPLY CHAIN (Number)', y='SUPREME', data=df, ax=axes[1, 1])
axes[1, 1].set_title('SUPREME')

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Display the plot
plt.show()


```
